### Explore Price Gaps
- Quantify a "large" price gap
- Identify when stocks are "close" to a price gap
- Evaluate performance of stock during those price gaps

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objects as go

symbol = "SMCI"
period = "1y"
df = yf.download(symbol, period = period)
df = df.reset_index()

fig = go.Figure(data=[go.Candlestick(x=df.index,
                open = df['Open'],
                high = df['High'],
                low = df['Low'],
                close = df['Close'])])

fig.update_layout(
    autosize=False,
    width=1000,
    height=800, 
    paper_bgcolor='black',
    plot_bgcolor='black')
fig.update_xaxes(gridcolor='black')
fig.update_yaxes(gridcolor='black')

fig.show()
```
